## Request를 이용해 웹페이지 다운

```python
import requests

url = 'https://www.yna.co.kr/sitemap/articles/2023/02/01-1.htm'
response = requests.get(url)

if response.status_code == 200:
    html = response.text
    print("Success")
    # print(html)
else:
    print("Error:", response.status_code)
```

## Beautiful Soup으로 파싱

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, "html.parser")
```

## 기사목록 태그 추려내기

```python
ul_tag = soup.find("ul",{"id":"sitemap-list"})
ul_tag
```

```python
for i in ul_tag.find_all('a'):
    print(i.text)
```

```python
[i.text for i in ul_tag.find_all('a')]
```

## 특정 일자의 연합뉴스 제목을 뽑아내는 함수를 만들어보시오

```python
def headline_list(month, date):
    url = f'https://www.yna.co.kr/sitemap/articles/2023/{month}/{date}-1.htm'
    response = requests.get(url)
    if response.status_code == 200:        
        html = response.text
        print("Success")
        soup = BeautifulSoup(html, "html.parser")
        ul_tag = soup.find("ul",{"id":"sitemap-list"})
        result = [i.text for i in ul_tag.find_all('a')]
        return result
    else:
        print("Error:", response.status_code)
        return None
    
```

```python
headline_list('02','04')
```

## 데이터베이스에 추가하기

```python
import sqlite3

conn = sqlite3.connect('news.db') ## 데이터베이스 접속하기
cursor = conn.cursor() # 마우스 커서처럼 데이터베이스의 현재 상태를 제어하는 도구

# 쿼리문 작성시 주로 사용됨
query = '''CREATE TABLE IF NOT EXISTS news(
            id INTEGER PRIMARY KEY,
            title TEXT           
)
'''
# 쿼리 실행
cursor.execute(query) #테이블이 만들어진다

```

```python
def create_news(title):
    query = f"INSERT INTO news (title) VALUES(?)"
    values = (title,)
    cursor.execute(query, values)
    conn.commit()
    print("Insert success")        
```

```python
def read_employee():
    query = "SELECT * FROM news"    
    cursor.execute(query)
    employees = cursor.fetchall()
    if employees:
        for record in employees:
            print(record)
    else:
        print("news not found.")
```

```python
news = headline_list('02','05')
for title in news:
    create_news(title)
```

```python
read_employee()
```

## Request를 이용하여 웹페이지 소스코드 다운받기

```python
# 외부 라이브러리인 request를 사용
import requests

url = "https://www.yna.co.kr/sitemap/" # 스크래핑 할 웹페이지의 주소
response = requests.get(url) # 웹페이지의 소스코드를 다운받아 변수에 저장

if response.status_code == 200: # 정상실행되었다면
    html = response.text # 소스코드를 html 변수에 저장
    print(html)
else: # 에러가 났다면
    print("Error:", response.status_code) #에러메시지 출력
```

```python
from bs4 import BeautifulSoup

html = """
<html>
  <head>
    <title>My Webpage</title>
  </head>
  <body>
    <h1>Welcome to my webpage</h1>
    <div class="content">
      <p>This is a paragraph.</p>
      <p>This is another paragraph.</p>
    </div>
  </body>
</html>
"""

soup = BeautifulSoup(html, "html.parser")

# h1 태그 내부의 텍스트 추출
h1_tag = soup.find("h1")
print(h1_tag.text)

# 첫 번째 p 태그 내부의 텍스트 추출
p_tag = soup.find("p")
print(p_tag.text)

# 모든 p 태그 내부의 텍스트 추출
p_tags = soup.find_all("p")
for p_tag in p_tags:
    print(p_tag.text)
```
